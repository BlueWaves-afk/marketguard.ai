```python
import yt_dlp

def download_audio(url):
    ydl_opts = {
        'format': 'bestaudio/best',
        'outtmpl': 'yt_audio.%(ext)s',
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
            'preferredquality': '192',
        }],
    }
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])

# Example: replace with any stock market YouTube link
download_audio("https://www.youtube.com/watch?v=PI709plS2uk")
```

```python
import ffmpeg
import numpy as np
from faster_whisper import WhisperModel

# Load a lightweight Whisper model (tiny = fastest, base = better balance)
model = WhisperModel("tiny.en", device="cpu")

def stream_transcribe(audio_file, chunk_size=5):
    """Simulate real-time transcription by processing audio in small chunks"""

    # Convert audio to raw PCM
    process = (
        ffmpeg
        .input(audio_file, threads=0)
        .output("pipe:", format="s16le", acodec="pcm_s16le", ac=1, ar="16000")
        .run_async(pipe_stdout=True, pipe_stderr=True)
    )

    chunk_bytes = 16000 * 2 * chunk_size  # 16k samples/sec * 2 bytes/sample * chunk_size
    buffer = b""

    i = 0
    while True:
        in_bytes = process.stdout.read(chunk_bytes)
        if not in_bytes:
            break

        buffer += in_bytes
        samples = np.frombuffer(buffer, np.int16).astype(np.float32) / 32768.0

        # Transcribe current chunk
        segments, _ = model.transcribe(samples, beam_size=3)
        print(f"\n--- Chunk {i} ---")
        for seg in segments:
            print(f"[{seg.start:.2f}s → {seg.end:.2f}s] {seg.text}")

        i += 1
        buffer = b""  # reset buffer after each chunk

# Run streaming transcription
stream_transcribe("yt_audio.mp3", chunk_size=5)
```
